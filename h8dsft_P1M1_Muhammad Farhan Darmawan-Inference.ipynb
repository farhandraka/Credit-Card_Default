# **XI. Model Inference**

## Import Libraries

```python
import pickle
import json
import pandas as pd
import numpy as np
```

## Load Model

```python
# Load Model

# Modelling
with open('ab_model.pkl', 'rb') as file_1:
    ab_model = pickle.load(file_1)


# Pre-processing

with open('scale_feat.pkl', 'rb') as file_2:
    scale_feat = pickle.load(file_2)

with open('winsoriser.pkl', 'rb') as file_3:
    winsoriser = pickle.load(file_3)
    
# List Numeric & Category
with open('num_cols_sc.txt', 'r') as file_4:
    num_cols_sc = json.load(file_4)
    
with open('num_cols_nsc.txt', 'r') as file_5:
    num_cols_nsc = json.load(file_5)
```

## Inferencing

```python
# Me-loading data
data = pd.read_csv('D:\RAKA\DATA SCIENCE\HACKTIVE8\Dataset GC\h8dsft_P1M1_Muhammad Farhan Darmawan.csv')
```

```python
df_inf = data.sample(n = 20)
df_inf
```

```python
# Mengambil data yang digunakan
df_inf = df_inf.drop(['default_payment_next_month'], axis = 1)
```

Handling Cardinality

```python
# Handling cardinality
df_inf = df_inf.replace({'education_level' : {1 : 1, 2 : 2, 3 : 3, 4 : 4, 5 : 4, 6 : 4, 0 :4}, 
                           'marital_status' : {1 : 1, 2 : 2, 3 : 3, 0 : 3},
                           'pay_0' : {-2 : 0, -1 : 0},
                           'pay_2' : {-2 : 0, -1 : 0},
                           'pay_3' : {-2 : 0, -1 : 0},
                           'pay_4' : {-2 : 0, -1 : 0},
                           'pay_5' : {-2 : 0, -1 : 0},
                           'pay_6' : {-2 : 0, -1 : 0}})
```

Handling Outlier

```python
df_inf_t = winsoriser.transform(df_inf)
```

Splitting data by list

```python
df_inf_sc = df_inf_t[num_cols_sc]
df_inf_nsc = df_inf_t[num_cols_nsc]
```

Feature Scaling

```python
df_inf_sc = scale_feat.transform(df_inf_sc)
```

```python
df_inf_sc = pd.DataFrame(df_inf_sc, columns = num_cols_sc)
```

```python
# Reset Index
df_inf_sc.reset_index(drop= True, inplace= True)
df_inf_nsc.reset_index(drop = True, inplace = True)
```

```python
df_final = pd.concat([df_inf_nsc, df_inf_sc], axis= 1)
```

```python
df_final = df_final.loc[:,['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6', 'limit_balance', 
                                     'pay_amt_1', 'pay_amt_2', 'pay_amt_3', 'pay_amt_4', 'pay_amt_5', 'pay_amt_6']]
```

Modeling

```python
y_pred_inf = ab_model.predict(df_final)
y_pred_inf
```
